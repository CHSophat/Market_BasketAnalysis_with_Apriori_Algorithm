# 01. Data Exploration

In this notebook, we perform Explanatory Data Analysis (EDA) on the grocery orders dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline
```

## 1. Load Data

```python
data_path = os.path.join("..", "data", "raw", "grocery_orders.csv")
df = pd.read_csv(data_path)
df.head()
```

```python
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

## 2. Analysis

```python
print(f"Unique items: {df['itemDescription'].nunique()}")
print(f"Unique members: {df['Member_number'].nunique()}")
```

```python
# Top 10 most frequent items
top_items = df['itemDescription'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_items.index, y=top_items.values, palette='viridis')
plt.title('Top 10 Best Selling Items')
plt.xticks(rotation=45)
plt.show()
```
